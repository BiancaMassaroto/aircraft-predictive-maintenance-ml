```python
import os
import pandas as pd

archives_input = "/Users/biancamassaroto/Documents/coding/ml_projects/aircraft_predictive_maintenance_ml/data/raw/archive"
archives_base_output = "Users/biancamassaroto/Documents/coding/ml_projects/aircraft_predictive_maintenance_ml/data/modified"

columns = ['unit_id', 'time_cycles', 'op_setting_1', 'op_setting_2', 'op_setting_3'] + \
          [f'sensor_{i}' for i in range(1, 22)]

for file in os.listdir(archives_input):
    print("Found file:", file)
    
    if "train" in file or "test" in file:
        file_path = os.path.join(archives_input, file)
        
        df = pd.read_csv(
            file_path,
            sep = r"\s+",
            header = None,
            names = columns
        )
        
        foulder_name = file.replace(".txt", "")
        archives_output = os.path.join(archives_base_output, foulder_name)
        
        os.makedirs(archives_output, exist_ok = True)
        
        new_file_name = f"{foulder_name}_processed.csv"
        output_path = os.path.join(archives_output, new_file_name)
        
        df.to_csv(output_path, index = False)
        print(f"Saved: {output_path}")
        
        
```
